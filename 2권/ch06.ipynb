## RNN의 문제점
기울기 소실과 기울기 폭발의 원인

```python
import numpy as np 
import matplotlib.pyplot as plt

N = 2
H = 3
T = 20

dh = np.ones((N, H))
np.random.seed(3)
Wh = np.random.randn(H, H)

norm_list = []
for t in range(T):
    dh = np.matmul(dh, Wh.T)
    norm = np.sqrt(np.sum(dh**2)) / N
    norm_list.append(norm)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(T), norm_list)
plt.xlabel('time step')
plt.ylabel('norm')
plt.xticks(np.array([1,5,10,15,20]))
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

N = 2
H = 3
T = 20

dh = np.ones((N, H))
np.random.seed(3)
Wh = np.random.randn(H, H) * 0.5

norm_list = []
for t in range(T):
    dh = np.matmul(dh, Wh.T)
    norm = np.sqrt(np.sum(dh**2)) / N
    norm_list.append(norm)

import matplotlib.pyplot as plt

plt.plot(range(T), norm_list)
plt.xlabel('time step')
plt.ylabel('norm')
plt.xticks(np.array([1,5,10,15,20]))
plt.show()
```
